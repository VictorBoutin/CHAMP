```python
class MonIter():
    
    def __init__(self,stop):
        self.current=0
        self.stop=stop
        
    def __iter__(self):
        return self
    
    def __next__(self):
        self.current+=1
        
        if self.current>self.stop:
           
            self.__init__(self.stop)
            raise StopIteration
        return self.current
```

```python
dico = dico_random
activation = torch.zeros(30)
res_list = list()
eta = 0.05
modulation = torch.ones(30)
for i_epoch in range(3):
    tic = time.time()
    for i,wrapped_data in enumerate(training_set) : 
        image,lable = wrapped_data
        #image = Variable(image)
        event, res, nb_activation, residual_patch = CodingCHAMP(image, dico, \
                                    l0_sparseness=10, modulation=modulation)
        accel = torch.mean(event[1])
        dico.data.add_(residual_patch.mul_(eta).view(dico.size()))
        dico = torch.mul(dico,mask)
        dico = Normalize(dico)
        res_list.append(res)
        activation += nb_activation
        modulation = UpdateModulation(modulation, activation, eta_homeo=0.01)
    print('epoch {0} done in {1:.2f} s'.format(i_epoch, time.time()-tic))
```
