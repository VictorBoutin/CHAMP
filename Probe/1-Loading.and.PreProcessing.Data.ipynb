```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
from CHAMP.DataLoader import LoadData
from CHAMP.DataTools import LocalContrastNormalization, FilterInputData
from CHAMP.Monitor import DisplayDico
# inform on the location of your dataset
# datapath = "../../../Science/These/Rapport d'avancement/database/"
import os
home = os.getenv('HOME')
datapath = os.path.join(home, "science/VB_These/Rapport d'avancement/database")
```

## Face Data Base (AT&T)

The function created to load the face database could be used to load any custom database from a specific folder. 

### Loading Raw Data

```python
path = os.path.join(datapath, "Face_DataBase")
```

```python
TrSet, TeSet = LoadData('Face', path, decorrelate=False, resize=(65, 65))
to_display = TrSet[0][0, 0:10, :, :, :]
print('Size=', TrSet[0].shape)
DisplayDico(to_display)
```

### Performing Local Contrast Normalization

```python
N_TrSet, _, _, _ = LocalContrastNormalization(TrSet)
to_display = N_TrSet[0][0, 0:10, :, :, :]
DisplayDico(to_display)
```

### Applying Gaussian Mask

```python
Filtered_L_TrSet = FilterInputData(
    N_TrSet, sigma=0.25, style='Custom', start_R=15)
to_display = Filtered_L_TrSet[0][0, 0:10, :, :, :]
DisplayDico(to_display)
```
